```python
import requests
from bs4 import BeautifulSoup
import time
```

```python
page = 1
books = []
headers = {'user-agent': 'Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/52.0.2743.116 Safari/537.36'}
while page < 3:
    url = "https://www.books.com.tw/web/sys_bbotm/books/010101/?o=1&v=1&page=" + str(page)
    re = requests.get(url, headers = headers)
    re.encoding='utf8'
    soup = BeautifulSoup(re.text,"html.parser")
    
    books_data = soup.select('.wrap')
    time.sleep(1)
    #抓取所有class為item的資料
    book_data = books_data[0].select('.item')
    print(book_data)
    for book in book_data:
        #抓取書名
        title = book.find('h4')
        print(title.text)
        #抓取作者、出版社、出版日期
        data = book.find('li').text.split('，')
        author = data[0]
        print(author)
        Publisher = data[1]
        published = data[2].split('：')[1]
        print(Publisher)
        print(published)
        #抓取價錢
        prices = book.find_all('strong')
        #有打折
        if(len(prices) > 1):
            price = prices[1].text
        #沒打折
        else:
            price = prices[0].text
        print(price)
        #抓取連結
        link = title.find('a')['href']
        print(link)
        #存入books串列
        books.append({"Author":author,"Tittle":title.text,"Publisher":Publisher,"published":published,"price":price,
                    "link":link})
    time.sleep(1)
    page +=1
```

```python
print(books)
```

```python
import pandas as pd

# Create a Pandas dataframe from the data.
df1 = pd.DataFrame(books)

# Create a Pandas Excel writer using XlsxWriter as the engine.
writer1 = pd.ExcelWriter('Books.xlsx', engine='xlsxwriter')#改檔名
# Convert the dataframe to an XlsxWriter Excel object.
df1.to_excel(writer1, sheet_name='Sheet1')
# Close the Pandas Excel writer and output the Excel file.
writer1.save()
```
